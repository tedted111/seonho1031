# 예제 문제

```python
import pandas as pd 
import numpy as np 
import matplotlib.pypl
```

```python
# 'exam_sample.csv' 파일 읽기  
df1 = pd.read_csv('exam_sample.csv')
df1
```

```python
# 'exam_sample01.csv' 파일 읽기
df2 = pd.read_csv('exam_sample01.csv')
df2
```

```python
# df1, df2 concat 후 result1으로 저장 : axis= 0
result1 = pd.concat([df1,df2], axis= 0)
result1
```

```python
# df1, df2 concat 후 result2로 저장: axis= 1
result2 = pd.concat([df1,df2], axis= 1)
result2
```

```python
# df1, df2 merge 후 저장 result3으로 저장 
result3 = pd.merge(df1, df2)
result3
```

```python
# df1을 'student_no'를 인덱스로 지정 후 df3로 저장 : set_index
df3= df1.set_index('student_no')
df3
```

```python
# df2를 'student_no'를 인덱스로 지정 후 df4로 저장 : set_index
df4= df2.set_index('student_no')
df4
```

```python
# df3, df4 concat 후 result4로 저장 : axis= 1
result4 = pd.concat([df3,df4], axis= 1)
result4
```

```python
# df3, df4 merge 후 저장 result5로 저장 : 'student_no' 기준으로
result5 = pd.merge(df3, df4, on='student_no')
result5
```

```python
# df3, df4 join(how='left') 후 저장 result6으로 저장
result6 = df3.join(df4, how='left')
result6
```

### 데이터 그룹 분석 

#### 집계합수

```python
# result2에서 'student_no'컬럼을 drop해서 df에 저장
df = result2.drop(columns=['student_no'])
df
```

```python
# df를 'class' 기준으로 groupby : 결측치를 제외한 표준편차 : std()
df.groupby('class').std(numeric_only=True)
```

```python
# df를 'gender' 기준으로 groupby : sum() 
result = df.groupby('gender').sum(numeric_only=True)
result
```

```python
# df를 'gender', 'class' 기준으로 groupby : mean() 
result = df.groupby(['gender', 'class']).mean(numeric_only=True)
result
```

```python
# class별 로 그룹화(A,B,C반)
# class별 그룹 오브젝트만 생성
df1=df.groupby(['class'])
df1 
```

```python
# df1에서 'class' 중 A반 그룹만 확인
df1.get_group('A')
```

```python
# df1에서 'class'별 그룹 평균 확인
df1.mean(numeric_only=True)
```

```python
# 'class'별 'math' 평균
#df['math'].groupby(df['class']).mean()
#df.groupby('class')['math'].mean()
df1['math'].mean()
```

```python
# 'class'별 'math' 개수
#df['math'].groupby(df['class']).count()
df1['math'].count()
```

```python
# 'gender'별 'math' 평균
df_mean = df['math'].groupby(df['gender']).mean()
df_mean
```

```python
# 남학생('m')의 'math'점수
group = df.groupby('gender')
male = group.get_group('m')
subset = male[['gender', 'math']]
print(male)
print(subset)
```
